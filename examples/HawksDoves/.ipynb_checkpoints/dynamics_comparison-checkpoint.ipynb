```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ternary
```

```python
data_path = 'results//data'
df1 = pd.read_csv(os.path.join(data_path,'replicator.csv'))
df2 = pd.read_csv(os.path.join(data_path,'logit.csv'))
df3 = pd.read_csv(os.path.join(data_path,'smith.csv'))
df4 = pd.read_csv(os.path.join(data_path,'qlearn.csv'))
df5 = pd.read_csv(os.path.join(data_path,'pgrad.csv'))
df6 = pd.read_csv(os.path.join(data_path,'ppo.csv'))
```

## Ternary Plots

### Replicator

```python
fig,tax = ternary.figure(scale=1.0)
tax.set_title('Replicator Dynamics\n\n')
tax.boundary(linewidth=0.4)
tax.top_corner_label('D')
tax.left_corner_label('H')
tax.right_corner_label('M')
tax.get_axes().axis('off')

for s0 in df1['s0'].unique():
    tax.plot(df1[df1['s0']==s0][['D','H','M']].values,linewidth=2.0,color='black')
    
tax.show()
```

## Results Analysis

```python
import plotly.graph_objs as go
import plotly.express as px
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
fig = go.Figure()

s1 = df1.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s1['D'],y=s1['H'],z=s1['M'],name='Replicator',mode='markers'))
s2 = df2.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s2['D'],y=s2['H'],z=s2['M'],name='Logit',mode='markers'))
s3 = df3.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s3['D'],y=s3['H'],z=s3['M'],name='Smith',mode='markers'))
s4 = df4.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s4['D'],y=s4['H'],z=s4['M'],name='Q-learning',mode='markers'))
s5 = df5.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s5['D'],y=s5['H'],z=s5['M'],name='PG',mode='markers'))
s6 = df6.groupby('s0').last()
fig.add_trace(go.Scatter3d(x=s6['D'],y=s6['H'],z=s6['M'],name='PPO',mode='markers'))

fig.show()
```
