```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ternary
```

```python
import sys
sys.path.append('..//..//')
import LDynamics
```

```python
e = LDynamics.Evaluator('RPS',strategies=['R','P','S'])
```

```python
e.evaluate()
```

```python
df1 = e.results[0][1]
df2 = e.results[1][1]
df3 = e.results[2][1]
df4 = e.results[3][1]
df5 = e.results[4][1]
```

### Replicator

```python
fig,tax = ternary.figure(scale=1.0)
tax.set_title('Replicator Dynamics\n\n')
tax.boundary(linewidth=0.4)
tax.top_corner_label('R')
tax.left_corner_label('P')
tax.right_corner_label('S')
tax.get_axes().axis('off')

for s0 in df1['s0'].unique()[:9]:
    tax.plot(df1[df1['s0']==s0].head(6000)[['R','P','S']].values,linewidth=2.0,color='black')
    
tax.show()
```

### Logit

```python
fig,tax = ternary.figure(scale=1.0)
tax.set_title('Logit Dynamics\n\n')
tax.boundary(linewidth=0.4)
tax.top_corner_label('P')
tax.left_corner_label('R')
tax.right_corner_label('S')
tax.get_axes().axis('off')

for s0 in df2['s0'].unique():
    tax.plot(df2[df2['s0']==s0].head(800)[['R','P','S']].values,linewidth=2.0,color='black')
    
tax.show()
```
